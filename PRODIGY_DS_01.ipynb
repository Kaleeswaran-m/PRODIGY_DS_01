In this task, I visualized the global population distribution using World Bank data. I explored the total population per country and analyzed which countries had the highest numbers.

### üì• Step 1: Importing Libraries

```python
# Importing basic libraries for data handling and visualization
import pandas as pd
import matplotlib.pyplot as plt
```

### üìÅ Step 2: Loading the Dataset
I downloaded the population data from the World Bank (indicator: SP.POP.TOTL). Here I load the CSV file and check a few initial rows.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the dataset (make sure you downloaded it and placed in same folder)
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_123525.csv', skiprows=4)
# Display first few rows to understand structure
df.head()
```

### üîé Step 3: Filtering Relevant Data

```python
# Extracting country name and 2022 population columns
df_filtered = df[['Country Name', '2022']].dropna()
# Sorting countries by population
top10 = df_filtered.sort_values(by='2022', ascending=False).head(10)
top10
```

### üìä Step 4: Bar Chart of Top 10 Populated Countries (2022)

```python
# Creating a bar chart for top 10 most populated countries
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2022'], color='darkred')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### üìâ Step 5: Histogram - Overall Population Distribution

```python
# Plotting a histogram to see how population is distributed across all countries
plt.figure(figsize=(8,5))
plt.hist(df_filtered['2022'], bins=30, color='blue')
plt.title('Population Distribution Across Countries (2022)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```

### üìù Final Observations
- Countries like China and India have significantly larger populations compared to others.
- The bar chart clearly highlights the top 10 most populated countries.
- The histogram shows that most countries have a relatively small population, with a few having very high numbers.
